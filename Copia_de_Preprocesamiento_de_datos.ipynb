### Descripción de área del problema
Las empresas de comida rápida ofrecen una variedad de productos que contienen distintos valores nutricionales, con la mayoría coincidiendo en tener una alta concentración en las calorías, grasas, y otros elementos. Por ello, existe un programa de nutrición conocido como WeightWatchers Points. El cual es un sistema de puntuación que toma en cuenta elementos como las grasas, proteínas, calorías y azúcares y mediante cálculos obtiene una puntuación cuyo valor representa qué tan saludable es dicho alimento. Entre menos puntos, menos insalubre.
Obtener este puntaje requiere un proceso de cálculo entre los elementos antes mencionados, y esto puede tomar algo de tiempo considerando nuevos productos alimenticios, por lo que se necesita una opción que automatice esta tarea.
Descripción de soluciones actuales
Una posible solución para este problema sería el uso de herramientas de aprendizaje automatizado, las cuales permiten predecir el valor de un atributo a partir de los valores de otros mediante la detección de patrones entre instancias.
La principal ventaja de esta solución es que permite realizar cálculos de manera rápida tomando grandes cantidades de registros e identificando posibles patrones. Aun así, una posible desventaja sería la variabilidad en la exactitud de las predicciones, obteniendo resultados que podrían ser inexactos o muy alejados de la realidad.
### Definición de objetivos de negocio
Uno de los objetivos de este proceso es el de realizar predicciones de WW Points dada la información de los atributos de un alimento. Estos atributos podrían ser la cantidad de grasas (Saturadas y Trans), el colesterol, el sodio, los carbohidratos, los azúcares, entre otros más.
Otro objetivo es lograr crear un modelo que le permita conocer a la gente que tan saludable es un alimento en base a los WW points calculado con las columnas clave que vamos a identificar en futuras fases del proyecto.

#Modelado
## Selección de técnicas de modelado
Lista de las técnicas que se utilizarán para la evaluación del modelo
En lugar de dividir los datos en un único conjunto de entrenamiento y prueba, se utilizará la técnica de K-Fold Cross Validation para realizar una evaluación más efectiva, especialmente considerando la cantidad de registros disponibles. Este método permite evaluar el éxito del modelo mediante la partición del conjunto de datos en K subconjuntos (folds) de tamaño similar.
En cada iteración, uno de los folds se usa como conjunto de prueba, mientras que los restantes K−1 folds se utilizan para entrenar el modelo. Antes de crear los folds, los registros serán ordenados aleatoriamente para garantizar una distribución más equilibrada y reducir posibles sesgos. Este proceso se repite K veces, asegurando que cada subconjunto actúe como conjunto de prueba una vez. Al final, se calculan las métricas de evaluación, como el RMSE (Root Mean Square Error), para cada iteración, y se obtiene un promedio que refleja la precisión general del modelo.
Además, se implementará el algoritmo K-Nearest Neighbors (KNN), probando valores de K desde 1 hasta 9, para determinar cuál de ellos ofrece los mejores resultados. Paralelamente, se aplicará regresión lineal simple como método comparativo. Esto
permitirá identificar cuál de los dos enfoques, KNN o regresión lineal, proporciona el mejor desempeño en términos de predicción y precisión de los datos.
### Descripción de las razones por las que fueron seleccionados dichos modelos
La razón principal es a causa de que la variable objetivo es numérica y por lo tanto identificamos que tenemos un problema de regresión.
Otra de las razones de la selección fue debido a uno de los modelos más empleados para resolver problemáticas de regresión. Otra razón es su fácil desarrollo y ejecución.

## Generación de plan de prueba
Se realizará la prueba de KNN con valores de K en un rango del 1 al 3, esto para comprobar los resultados con distinta cantidad de K, para identificar cuál valor de K nos brinda el mejor resultado del RMSE.
Indicar los números de iteraciones y técnicas de validación a ser utilizadas (por cada conjunto de datos)
Como se especificó en el punto anterior, se probará el algoritmo KNN un total de 3 veces, para comparar los distintos resultados del RMSE obtenidos de los diferentes valores de K para seleccionar el que tenga el RMSE mejor.
Buscamos un RMSE de máximo 10.
## Construcción de modelos
* Selección de parámetros: Indicar los parámetros seleccionados para cada uno de los modelos. Esto por cada modelo y conjunto de datos utilizado, si se realizan varias pruebas (diferentes selecciones de parámetros), indicar los parámetros para cada prueba realizada
La variable objetivo para los modelos es la de WeightWatchers, solo contamos con un dataset y probaremos 9 veces el algoritmo de KNN (con rangos de k de 1 a 9) y 9 veces el algoritmo de Regresión Lineal. Se sacará el promedio de los valores de RMSE que nos de cada algoritmo y con eso cual es el que nos da mejor resultado para nuestro RMSE.


```python
import pandas as pd
import random
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'

#Funcion para leer el csv de auto-mpg
def leer_csv2():
  df = pd.read_csv('WeightWatchersPts.csv')
  return df

#Funcion para generar el K-fold
def K_fold_KNN(df, k_Fold, k):
    # Número de instancias
    noInstancias = df.shape[0]

    # Creamos los índices y los combinamos aleatoriamente
    indices = np.arange(noInstancias)
    np.random.shuffle(indices)

    # Aplicamos K-Fold para generar índices para el entrenamiento y prueba.
    tamaño_fold = noInstancias // k_Fold
    resto = noInstancias % k_Fold
    inicio = 0

    if k_Fold == 1:
      #print("\n CASO DE SOLO UN FOLD")

      #Creamos el conjunto de entrenamiento y el de prueba con todo el conjunto
      entrenamiento_df = df
      prueba_df = df

      RMSC = KNN(k, entrenamiento_df, prueba_df)

      return RMSC

    else: # EN CASO DE QUE K-FOLD SEA MAYOR DE UNO
      RMSCS = []
      for fold in range(k_Fold):
        print(f"\n### Fold {fold + 1} ###")

        # Definimos el tamaño de cada fold
        tamaño_actual_fold = tamaño_fold + 1 if fold < resto else tamaño_fold

        # Índices para este fold
        test_indices = indices[inicio:inicio + tamaño_actual_fold]
        train_indices = np.concatenate((indices[:inicio], indices[inicio + tamaño_actual_fold:]))  # Entrenamiento

        # Particionamos el conjunto de datos
        entrenamiento_df = df.iloc[train_indices]
        prueba_df = df.iloc[test_indices]

        #print(entrenamiento_df)
        #print(prueba_df)

        RMSCS.append(KNN(k, entrenamiento_df, prueba_df))

        inicio += tamaño_actual_fold

        print(f"\n### Termina Fold {fold + 1} ###")

      RMSC_promedio = sum(RMSCS) / len(RMSCS)

      return RMSC_promedio

#Función para el archivo de auto-mpg
def KNN(k, entrenamiento_df, prueba_df):
    #Inicializamos el Root Mean Square Error
    RMSC = 0
    df = leer_csv2()
    #df2 = df.copy()
    #entrenamiento_df = df2.sample(frac=0.7)
    #prueba_df = df2.drop(entrenamiento_df.index)

    #entrenamiento_df, prueba_df = K_fold(df2)

    #Si excluye la columna clase, vuelve a recorrer el dataframa para convertir
    #la columna mpg en númericos para no tener problemas con los tipos de datos en las operaciones

    for column_name in df.columns:
        if column_name == "Weight Watchers Pnts":
            columna_clase = column_name
            entrenamiento_df[column_name] = pd.to_numeric(entrenamiento_df[column_name], errors='coerce', downcast='float')
            prueba_df[column_name] = pd.to_numeric(prueba_df[column_name], errors='coerce', downcast='float')
            continue
        else:
            # Convertimos a númericos y normalizamos las columnas restantes en las tablas de entrenamiento y de prueba
            entrenamiento_df[column_name] = pd.to_numeric(entrenamiento_df[column_name], errors='coerce', downcast='float')
            val_min = entrenamiento_df[column_name].min()
            val_max = entrenamiento_df[column_name].max()
            entrenamiento_df[column_name] = entrenamiento_df[column_name].map(lambda x: (x - val_min) / (val_max - val_min))

            prueba_df[column_name] = pd.to_numeric(prueba_df[column_name], errors='coerce', downcast='float')
            val_min = prueba_df[column_name].min()
            val_max = prueba_df[column_name].max()
            prueba_df[column_name] = prueba_df[column_name].map(lambda x: (x - val_min) / (val_max - val_min))

            #Se crean dos arreglos para guardar las distancias de cada instancia de prueba,
            #Es decir se saca la distancia de cada instancia de prueba por todas las instancias de entrenamiento
            distancias = []
            for idx_prueba, instancia_prueba in prueba_df.iterrows():
                distancia_instancia = {}
                for idx_entrenamiento, instancia_entrenamiento in entrenamiento_df.iterrows():
                    suma = 0
                    for column_name in entrenamiento_df.columns:
                        if column_name == "Calories":
                            continue
                        suma += (instancia_prueba[column_name] - instancia_entrenamiento[column_name]) ** 2
                    distancia = np.sqrt(suma)
                    distancia_instancia[idx_entrenamiento] = distancia  # Almacenar en el diccionario
                distancias.append(distancia_instancia)
    #Obtenemos los resultados de todas las distancias de entrenamiento por cada instancia de prueba
    MSC = 0
    for idx_prueba, distancias_instancia in zip(prueba_df.index, distancias):
        # print(f"Distancias de la instancia de prueba {idx_prueba}:")

        # Se obtienen los "k" vecinos más cercanos o resultados minimos para cada instancia de prueba
        k_minimos = sorted(distancias_instancia.items(), key=lambda x: x[1])[:k]
        # print(f"{k} valores mínimos para {idx_prueba}: {k_minimos}")

        #Este es un arreglo que almacena los "k" resultados minimos de las distancias de entrenamiento,
        #por cada instancia de prueba
        distancia_cercana = []
        for idx_entrenamiento, _ in k_minimos:
            class_value = entrenamiento_df.loc[idx_entrenamiento, columna_clase]
            distancia_cercana.append(class_value)
        # print(f"Valores cercanos de la clase para {idx_prueba}: {distancia_cercana}")

        #Sacamos el promedio de las clases que aparecen segun el arreglo de resultados minimos
        #Y mostramos que valor clase aparece en la instancia de prueba por cada arreglo de resultados
        #Es decir el promedio es "None" y la clase que aparece en su respectiva instancia de prueba es "Hard"
        if k > 0:
            promedio_clase = sum(distancia_cercana) / k
        else:
            promedio_clase = 0
        # print(f"Promedio de clase para {idx_prueba}: {promedio_clase}")

        valor_clase_prueba = prueba_df.loc[idx_prueba, columna_clase]
        # print(f"Valor actual de la instancia de prueba {idx_prueba} en columna de clase: {valor_clase_prueba}\n")
        #Obtenemos el Error Cuadratico Medio y la raiz del Error
        MSC += (promedio_clase - valor_clase_prueba) ** 2
    MSC = MSC / len(prueba_df)
    RMSC = np.sqrt(MSC)


    print(f"Error Cuadratico Medio: {MSC}")
    print(f"Raiz del Error Cuadratico Medio: {RMSC}")
    return RMSC

def main():
  #En los arreglos se almacenan los resultados finales de cada función
  k_lista_lenses = []
  k_lista_mpg = []

  min = 1000
  max = 0
  #Se ejecutan k numeros de veces el programa y con ese mismo número se aplican los algoritmos, es decir se obtendran k-vecinos más cercanos
  #dependiendo el valor que tome el numero en cada iteración
  #Por ejemplo en la primera iteración k vale 1, en la segunda iteracion k vale 2
  #Se le sumara uno por cada iteración que hagan con las funciones
  #Los resultados finales de dichas iterraciones se almacenan en arreglos

  df = leer_csv2()
  df2 = df.copy()

  for i in range(1,4):
    print(f"\n##### Valor de K = {i} #####")
    k_lista_mpg.append(K_fold_KNN(df2, 3, i))
    print(f"\nError cuadratico medio promedio: {k_lista_mpg[-1]}")
  print(k_lista_mpg)

  k_mpg = 1
  k=0

  for i in k_lista_mpg:
    k += 1
    if i < min:
      min = i
      k_mpg = k


  print(f"El valor minimo del Root Mean Square Error (RMSE) en el conjunto de datos de mpg es: {min} y se obtuvo con un valor de k = {k_mpg}")

  #print(f"El Root Mean Square Error del conjunto de datos de prueba es: {RMSC}")
  #print(f"El Root Mean Square Error del conjunto de datos de prueba es: {RMSC}")

if __name__ == '__main__':
  main()
```
